```python
import torch
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader

from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
trans = transforms.Compose([  # transform을 한번에 하게 해줌.
    transforms.Resize((64,128))  # 256 x 512는 너무 사이즈가 커서 줄인다.
]) 
train_data = torchvision.datasets.ImageFolder(root='C:/Users/hojun/Desktop/Jun/Study_Note/Pytorch_Tutorial/Dataset/custom_data/origin_data', 
                                              transform=trans)
```

```python
for num, value in enumerate(train_data):  # enumerate가 쓰이면 num은 몇번째 반복문인지를 알려주는 값과 enumerate()안에있는 원소값을 튜플형태로 반환함
                                           # value에는 train_data의 원소. 즉 이미지와 라벨이 동시에 들어가 있고 이건 value 변수안에 저장됨
                                            # num에는 단지 몇번째 반복문인지에 대한 숫자만 저장되어 있음
    data, label = value
    print(num, data, label)
    
    if(label==0): # 라벨은 gray, red 단 두개이므로 0, 1 두개로 구분한다
        data.save('C:/Users/hojun/Desktop/Jun/Study_Note/Pytorch_Tutorial/Dataset/custom_data/train_data/gray/%d_%d.jpeg'%(num, label))
    else:
        data.save('C:/Users/hojun/Desktop/Jun/Study_Note/Pytorch_Tutorial/Dataset/custom_data/train_data/red/%d_%d.jpeg'%(num, label))
    imshow(data)
```
